```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# criando uma varíavel com distribuição normal (1000 elementos)
dados = norm.rvs(size = 1000)
dados
```

```python
# histograma para visualizar a distribuição
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
# gerando um qqplot para verificar se a distribuição é normal
fig, ax = plt.subplots()
stats.probplot(dados, fit= True, plot= ax)
plt.show()
```

```python
# executando o teste de shapiro
stats.shapiro(dados)
```

```python
# criando uma distribuição não normal
dados2 = skewnorm.rvs(4, size = 1000)
```

```python
# fazendo um histograma para visualizar
plt.hist(dados2, bins = 20)
plt.title('dados2')
```

```python
# qqplot
fig, ax = plt.subplots()
stats.probplot(dados2, fit= True, plot= ax)
plt.show()
```

```python
# teste de shapiro
stats.shapiro(dados2)
```
