```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('co2.csv')
base.head()
```

```python
srn.scatterplot(base.conc, base.uptake, hue=base.Type) # gráfico de dispersão agrupado pelo Type
```

```python
# filtrando dados específicos
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
# fazendo gráfiocos separados para cada região
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
srn.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
# novo filtro
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(ch.conc, ch.uptake).set_title('chilled')
plt.subplot(1,2,2)
srn.scatterplot(nc.conc, nc.uptake).set_title('nonchilled')
plt.tight_layout()
```
